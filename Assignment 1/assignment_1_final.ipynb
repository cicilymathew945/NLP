### 2. Use regular expressions to extract the regulation description and inspector comments from the "violations" columns.

```python
def extract_violations(text):
    """
    Extracts violations from a given text string.

    This function parses a string containing violation details, splits it into individual violations,
    and extracts the code, description, and optional comments for each violation.

    Parameters:
    text (str): A string containing violation details, where each violation is separated by '|'.
                Each violation is expected to follow the format:
                '<code>. <description> - Comments: <comment>'.

    Returns:
    list: A list of tuples, where each tuple contains:
          - code (str): The violation code.
          - desc (str): The violation description.
          - comment (str): The optional comment associated with the violation. If no comment is present, an empty string is returned.
    """
    if pd.isnull(text):
        return []
    
    violations = text.split('|')
    result = []
    for v in violations:
        v = v.strip()
        match = re.match(r'^\s*(\d+)\.\s*(.*?)(?:\s*-\s*Comments:\s*(.*))?$', v, re.IGNORECASE)
        if match:
            code = match.group(1).strip()
            desc = match.group(2).strip()
            comment = match.group(3).strip() if match.group(3) else ''
            result.append((code, desc, comment))
    return result
```

```python
# Parsing the violations
df_failed["Parsed_Violation"] = df_failed["Violations"].apply(extract_violations)
df_failed = df_failed.explode("Parsed_Violation")
```

### 4. Tokenize the comments and provide a count of the top 10:
   1. Raw tokens
   2. Preprocessed tokens
   3. Preprocessed tokens + Porter stemmer
   4. Preprocessed tokens + Lancaster stemmer
   5. Preprocessed tokens + lemmatization

```python
porter = nltk.PorterStemmer()
lancaster = nltk.LancasterStemmer()
lemmatizer = nltk.WordNetLemmatizer()
```

```python
def get_wordnet_pos(treebank_tag, skip_value = "n"):
    """
    Converts treebank part-of-speech tags to WordNet part-of-speech tags.

    This function maps the treebank POS tags (used in NLTK's pos_tag function) 
    to the corresponding WordNet POS tags, which are required for lemmatization.

    Parameters:
    treebank_tag (str): The POS tag from the treebank (e.g., 'NN', 'VB', 'JJ', etc.).
    skip_value (str): The default WordNet POS tag to return if no match is found. Default is 'n' (noun).

    Returns:
    str: The corresponding WordNet POS tag (e.g., 'a' for adjective, 'v' for verb, 'n' for noun, 'r' for adverb).
    """
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return skip_value
```

```python
# cleaning
comments = df_failed["Violation_Comment"].dropna()
comments = comments[comments.apply(lambda x: isinstance(x, str))]
text = ' '.join(comments)
```

```python
raw_tokens = word_tokenize(text)
```

```python
# Filtering out non-alphabetic tokens and stop words
stop_words = set(stopwords.words('english'))
preprocessed = [
    token.lower()
    for token in raw_tokens
    if token.isalpha() and token.lower() not in stop_words
]
```

```python
porter_stemmed = [porter.stem(token) for token in preprocessed]
```

```python
lancaster_stemmed = [lancaster.stem(token) for token in preprocessed]
```

```python
tagged = pos_tag(preprocessed)
lemmatized = [lemmatizer.lemmatize(token, get_wordnet_pos(pos)) for token, pos in tagged]
```

```python
tokenized_component_names = [
    "Raw Tokens",
    "Preprocessed",
    "Porter Stemmed",
    "Lancaster Stemmed",
    "Lemmatized"
]

tokenized_components = [
    raw_tokens,
    preprocessed,
    porter_stemmed,
    lancaster_stemmed,
    lemmatized
]
```

```python
# Create a DataFrame to rank and print the top 10 tokens for each method
ranked_columns = {}

for name, tokens in zip(tokenized_component_names, tokenized_components):
    top_tokens = Counter(tokens).most_common(10)
    ranked_columns[name] = [f"{token}" for token, count in top_tokens]

ranked_df = pd.DataFrame(ranked_columns)
ranked_df.index += 1
ranked_df.index.name = "Rank"

ranked_df
```

```python
# Visualizing the top 10 tokens for each method
fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(25, 10), sharey=True)
fig.suptitle("Top 10 Tokens by Text Preprocessing Type", fontsize=16)

formatter = FuncFormatter(lambda x, _: f'{int(x / 1000)}K')  # Format x-axis in thousands

for ax, name, tokens in zip(axes, tokenized_component_names, tokenized_components):
    top = Counter(tokens).most_common(10)
    labels, counts = zip(*top)
    ax.barh(labels, counts)
    ax.set_title(name)
    ax.invert_yaxis()
    ax.xaxis.set_major_formatter(formatter)  # Apply formatter to x-axis

plt.tight_layout(rect=[0, 0, 1, 0.95])  # Leave space for suptitle
plt.show()
```
